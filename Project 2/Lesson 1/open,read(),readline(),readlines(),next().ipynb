```python
import os
```

```python
data_dir = ""
data_name = "beatles-diskography.csv"
data_file = os.path.join(data_dir,data_name)

# open()
f = open(data_file)
print type(f)
print f
#for i in f:
#    print i, type(i)
f.close()
```

```python
# Read()
f = open(data_file)
read_in = f.read()
print type(f.read()) is str
print read_in
for i in read_in:
    print i
f.close()
```

```python
# readline()
f = open(data_file)
print type(f.readline())
print f.readline()
print f.readline()   # string
print f.readline().split(',')
print type(f.readline().split(','))  # list
for i in f.readline():
    print i, type(i)
f.close()
```

```python
# Readlines()
f = open(data_file)
f_read = f.readlines()
print type(f_read)
for i in f_read:
    print i, type(i)
f.close()
```

```python
# next()
f = open(data_file)
f_line = f.next()
print type(f_line)
for i in f_line:
    print i, type(i)
f.close()
```
